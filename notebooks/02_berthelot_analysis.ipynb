# Berthelot analysis

```python
import pandas as pd
```

```python
# load gtf 
data_path ="/home/luisasantus/Desktop/crg_cluster/data/"
gtf_file = data_path + "CrossEvoPred/berthelot/Homo_sapiens/annotation/Homo_sapiens.GRCh37.73.gtf"
gtf = pd.read_csv(gtf_file, sep="\t", comment="#", header=None)
colnames = ["chr", "type", "subtype", "start", "stop", "x", "strand", "y", "meta"]
gtf.columns = colnames
```

```python
# fpkm 
fpkm_file = data_path + "CrossEvoPred/berthelot/Homo_sapiens/fpkm/do3239_unk_liver_unk_hsa-hg19Homo_sapiensCRI3_CRI01_20M.tophat.cufflinks_genes.fpkm"
fpkm = pd.read_csv(fpkm_file, sep = "\t")
fpkm
```

```python
# convert fpkm to bed 
# split locus 
fpkm["chr"] = fpkm.locus.str.split(":", expand = True)[0]
fpkm["pos"] = fpkm.locus.str.split(":", expand = True)[1]
fpkm["start"] = fpkm.pos.str.split("-", expand = True)[0]
fpkm["end"] = fpkm.pos.str.split("-", expand = True)[1]
```

```python
# keep columns gene_id, chr, start, end
bed = fpkm[["chr", "start", "end", "gene_id", "FPKM"]]

# write first ten lines of bed file 
bed_temp = bed[1:10]
bed_temp.to_csv('output.bed', index=False, sep = "\t", header = None) 
```
